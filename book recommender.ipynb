```python
def year(p,q,r,s):
    for i in range(len(p)):
        if q[i]=='DK Publishing Inc':
            author=p[i].split(';')[1][:-1]
            year=r[i]
            pub=q[i]
            r[i]=str(author)
            q[i]=str(year)
            s[i]=str(pub)
            p[i]=str(p[i].split('(')[0][:-1])
            
        elif q[i]=='Gallimard':
            author=p[i].split(';')[1][:-1]
            year=r[i]
            pub=q[i]
            r[i]=str(author)
            q[i]=str(year)
            s[i]=str(pub)
            p[i]=str(p[i].split('\'')[0][:-1])
        
    return np.array(p),np.array(r),np.array(s),np.array(q)
```

```python
def loc(x):
    a,b,c=[],[],[]
    for i in range(len(x)):
        p=x[i].split(',')
        if p[0]=='n/a':
            p[0]=np.NaN
        if p[1]==' n/a':
            p[1]=np.NaN
        if p[2]==' n/a':
            p[2]=np.NaN
        if p[1]==' ':
            p[1]=np.NaN
        if p[2]=='':
            p[2]=np.NaN
        a.append(p[0])
        b.append(p[1])
        c.append(p[2])
    return np.array(a),np.array(b),np.array(c)
```

### POPULARITY BASED

```python
new_df=df[['Book-Title','Book-Rating']]
new_df
```

```python
rating_count=new_df.groupby('Book-Title').count()
rating_count.rename(columns={'Book-Rating':'num_rating'},inplace=True)
rating_count
```

```python
rating_avg=new_df.groupby('Book-Title').mean()
rating_avg.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
rating_avg
```

```python
popular=rating_count.merge(rating_avg,on='Book-Title')
popular
```

```python
top_50=popular[popular['num_rating']>300].sort_values('avg_rating',ascending=False).head(50)
top_50
```

```python
top_50.merge(df,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']].reset_index(drop = True)
```

### COLLABORATIVE FILTERING BASED

```python
df_new=df[['Book-Title','Book-Rating','User-ID']]
df_new
```

```python
x=df_new.groupby('User-ID').count()['Book-Rating']>200
trusted_users=x[x].index
trusted_users
```

```python
filtered_users=df_new[df_new['User-ID'].isin(trusted_users)]
filtered_users
```

```python
y=filtered_users.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
famous_books
```

```python
final_rating=filtered_users[filtered_users['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_scores=cosine_similarity(pt)
sim_scores
```

```python
def recommend(book_name):
    index=np.where(pt.index=='1984')[0][0]
    similar=sorted(list(enumerate(sim_scores[0])),key=lambda x:x[1],reverse=True)[1:6]
        
    data = []
    for i in similar:
        item = []
        temp_df = df[df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
